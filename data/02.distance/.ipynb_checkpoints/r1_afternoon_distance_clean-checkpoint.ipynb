```python
raw_data
```

```python
raw_data['rtime'] = raw_data['date'] + " " + raw_data['time']
raw_data
```

```python
merge_data = raw_data[['rtime', 'distance']]
merge_data
```

```python
merge_data.isnull().sum()
```

```python
grouped = merge_data['distance'].groupby(merge_data['rtime'])
data_1 = grouped.min()
data_1
```

```python
# type(data_1)
```

```python
data = pd.DataFrame(data_1, columns=['distance'])
data
```

```python
data.reset_index(level=0, inplace=True)
data
```
