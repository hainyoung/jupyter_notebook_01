```python
gps_data_concat = pd.concat([gps_0, gps_2])
gps_data_concat
```

```python
gps_sorted = gps_data_concat.sort_values(by='rtime', ascending=True)
```

```python
gps_sorted
```

```python
gps_sorted.duplicated(['rtime']).sum()
```

```python
gps_data = gps_sorted.drop_duplicates('rtime', keep='first')
gps_data
```

```python
gps_data = gps_data[['rtime', 'latitude', 'longtitude']]
gps_data
```

```python
gps_data.to_csv('./merged/r1_night_gps_merged.csv', encoding='utf-8', index=False)
```
