```python
import numpy as np
import pandas as pd
```

```python
df_dash = pd.read_csv('./r1_night_inside_clean.csv', encoding='utf-8')
df_dash
```

```python
df_dist = pd.read_csv('./r1_night_distance.csv', encoding='utf-8')
df_dist
```

```python
df_gps = pd.read_csv('./r1_night_gps_merged.csv', encoding='utf-8')
df_gps
```

```python
df_merged_1 = pd.merge(df_dash, df_gps, on='rtime', how='outer')
df_merged_1
```

```python
df_merged_1.isnull().sum()
```

```python
data_1 = df_merged_1.interpolate()
data_1.isnull().sum()
```

```python
data = pd.merge(df_dist, data_1, on='rtime')
data
```

```python
data = data[['rtime', 'lkas', 'ldws', 'distance', 'speed', 'latitude', 'longtitude']]
data
```

```python
data.isnull().sum()
```

```python
data.to_csv('./final_r1_night.csv', encoding='utf-8', index=False)
```
